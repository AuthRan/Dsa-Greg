Sliding Window is a famous algo to solve problems. It has two types depending upon the window as variable length which means ur window has a variable length or it can change size. The second type is Fixed length sliding window meaning ur window has a fixed size like k=3.

For any sliding window problem you are given an array or string and you find keywords in question as Subarray or substring then its a huge give away that using sliding window might help.

Variable length sliding window problem Leetcode 3.

In sliding window we take two pointers L and R and whatever is btw the L and R is called our window. We slide R first till it is "valid" then slide L as well till we reach the end. Length of current window = R - L + 1


If we have valid window we move R and if we have invalid Substring we move L.
