Stacks in computer science can be thought of building left to right. Even though it looks like an array and infact it is an array but the whole point of stack is that you only care about what's on the top or in case of computer science what's on the right side only.

You can implement a stack either using an array where the cost of appending would we amortized cost *O(1) or you can implement it using a linked list where if you used doubly linked list then you will have access to tail pointer which will make the cost of appending O(1).

Another operation is to pop the topmost or rightmost (in computer science) its cost would be O(1) and similarly for peek operation cost is always O(1). Stacks are LIFO. It has an important function called isEmpty() -> in python written as if s: True

Queues on the other end are polar opposite of stacks which follows FIFO property. It has few different properties like stack's isEmpty like Enqueue (pronounce 'aunqueue') which simply means put in a new number x in queue which is similar to append operation in stack as well with amortized cost of *O(1). Another important operation is Dequeue (pronounce 'de-queue') which means to pop the first element in your data structure.

If queue is implemented using an array then the cost of dequeue would be O(n) which is large so in order to fix it we mostly implement Dequeue using a Doubly linked list which costs only O(1) to shift the head pointer and also the enqueue operation would cost only O(1) (not amortized O(1)). Because of this whenever you think of queue always think of doubly linked list.

Stacks -> dynamic array
Queue -> doubly linked list

```python
# Stack - Last in first out policy(LIFO)
stk = []
stk
```

```python
# Append to top of stack - O(1)
stk.append(4)
stk
```

```python
stk.append(8)
stk.append(3)
stk
```

```python
# Pop from stack - O(1)
x = stk.pop()
x
```

```python
# Peek at the top of stack - O(1)
print(stk[-1])
```

```python
# Check if stack is empty or not? - O(1)
if stk:
    print(True)
```

```python
# If you don't have any element and you try to pop then you will get an error so that's why before running the pop you should run the upper code ie to check if stack is not empty.
```

```python
# Queues - First in first out policy(FIFO)
# For simplifying things we use dequeue from collections
from collections import deque
queue = deque()
queue
```

```python
# Question is why it is called deque that's because it is a double ended queue ie you can push pop at both end front and back and its cost would be O(1) only.
# For our queue pourpose we will use it as a queue
```

```python
# Enqueue - Add an element to the right - O(1)
queue.append(5)
queue.append(15)
queue.append(50)
queue
```

```python
# Dequeue - Deletes an element from the left - O(1)
x = queue.popleft()
x
```

```python
# Peek from the left and peek from the right side - O(1)
left = queue[0]
right = queue[-1]
print("Left: ", left)
print("Right: ", right)
```
