A binary tree in leetcode is usually implemented using a class Node with node.value, node.left and node.right. There is something called complete binary tree in which nodes at all levels are filled and in those where it is not completely filled have nodes filled from left to right.

The binary tree can be represented using an array like this [1, 2, 3, 4..] for for each element i in this 2i is its left child and 2i+1 is its right child. There is also something called height assosicated with a binary tree ie logN

In binary tree we are often focussed on how we traverse the tree there can deapth first approach or breadth first approach. DFS can be of three types as given below, 

Another approach for traversal can be Breadth first search which is level order traversal unlike DFS. Another interesting point to remember is that DFS uses a stack as it is recurssive process whearas BFS uses a queue.

You can also implement dfs using iteration which gives rise to iterative dfs. Iterative just means without recurssion. 
The basic way to handle both bfs and dfs is that start with an empty stack or queue with root node in it then while stack or queue is not empty keep doing smthing.


The time and space complexity to search for a node in the binary tree is O(n) both. Either u use DFS or BFS the complexity remains the same.

The is something called Binary Search Tree which simply means that for any given node all the elements on its left are smaller than it and all the elements in its right are bigger than it. The great thing about BST is that it reduces the search cost by half bcz of which the cost of searching an element in BST is O(logn). As for any node T if x is less than it then all nodes on left of T are discarded viceversa for right side if T>x. But this works only when the height of BST is balanced if it is skewed on one side like 1->2->3->4.. soo on then searching again will cost O(n). So BST has a searching complexity of O(logn) iff it is height balanced

Also always remmeber that if you do inorder traversal (DFS) on BST then you will acesss all the elements in order (ie increasing order).

```python
# Binary Trees
class TreeNode:
  def __init__(self, val, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

  def __str__(self):
    return str(self.val)
```

```python
#        1
#     2    3
#   4  5  10
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
```

```python
# Recursive Pre Order Traversal (DFS) Time: O(n), Space: O(n)
def pre_order(node):
  if not node:
    return

  print(node)
  pre_order(node.left)
  pre_order(node.right)

pre_order(A)
```

```python
# Recursive In Order Traversal (DFS) Time: O(n), Space: O(n)
def in_order(node):
  if not node:
    return

  in_order(node.left)
  print(node)
  in_order(node.right)

in_order(A)
```

```python
# Iterative Pre Order Traversal (DFS) Time: O(n), Space: O(n)
def pre_order_iterative(node):
  stk = [node]

  while stk:
    node = stk.pop()
    if node.right: stk.append(node.right)  # Notice how we put right first in stack then left
    if node.left: stk.append(node.left)    # Because as this is a stack we need to pop rightmost.
    print(node)

pre_order_iterative(A)
```

```python
# Level Order Traversal (BFS) Time: O(n), Space: O(n)
from collections import deque

def level_order(node):
  q = deque()
  q.append(node)

  while q:
    node = q.popleft()
    print(node)
    if node.left: q.append(node.left)
    if node.right: q.append(node.right)

level_order(A)
```

```python
# Check if Value Exists (DFS) Time: O(n), Space: O(n)
def search(node, target):
  if not node:
    return False

  if node.val == target:
    return True

  return search(node.left, target) or search(node.right, target)

search(A, 11)
```

```python
# Binary Search Trees (BSTs)

#       5
#    1    8
#  -1 3  7 9

A2 = TreeNode(5)
B2 = TreeNode(1)
C2 = TreeNode(8)
D2 = TreeNode(-1)
E2 = TreeNode(3)
F2 = TreeNode(7)
G2 = TreeNode(9)

A2.left, A2.right = B2, C2
B2.left, B2.right = D2, E2
C2.left, C2.right = F2, G2

print(A2)
```

```python
# Time: O(log n), Space: O(log n)
def search_bst(node, target):
  if not node:
    return False

  if node.val == target:
    return True

  if target < node.val: return search_bst(node.left, target)
  else: return search_bst(node.right, target)

search_bst(A2, -1)
```
