Recurssive Backtracking is a famous algorithm. And as name suggest its going to be recurssive and we apply recurssion in such a way that it backtracks like you make a 1.decision then 2.by recurssion 3.you eventually hits a base case 4.then from which u undo your decisions.

Why we do all this ? Its generally to do an exhaustive search basically like a brute force technique. Usually like when you see "give all solution" this "all" means we might use backtracking to generate all the solutions.
