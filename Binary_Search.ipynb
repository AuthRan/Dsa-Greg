We can deal with it in two different ways ie Traditional method and over under technique(commonly seen in leetcode) 
So binary searched is applied on a sorted array like in traditional binary search you try to reduce the cost of linear search ie O(n) by taking two pointers one at left and other at right and try to find the middle of them if the N is even then you will get two middle then in that case you choose the lower index value the formula for middle M = (l+r) // 2
Then we check if our array[m] == target if not then check wheter it is less than or greater than array[m] if greater than then change l to m+1 or if less then change r to m-1 and keep doing this. This thing stops if l>r.
Now there is something called integer overflow which usually occurs when your both l and r get's too much right causing l+r to be very costly resulting in an integer overflow. To tackle this problem we use new formulae which is more futureproof M = L + [(R-L) // 2].

Another use is called over under method or conditional method for which take an example suppose we have to find from which day is he not sick [sick, sick, sick , not_sick, not_sick] then also we can implement binary search and everything including pseudocode is quite same.


Time complexity: logN

Space complexity: 1

```python
A = [-3, -1, 0, 1, 4, 7]

# Naive O(n) searching
if 1 in A:
  print(True)
```

```python
# Traditional Binary Search - Looking up if number is in array:
# Time: O(log n)
# Space: O(1)

def binary_search(arr, target):
  N = len(arr)
  L = 0
  R = N - 1

  while L <= R:
    M = L + ((R-L) // 2)

    if arr[M] == target:
      return True
    elif target < arr[M]:
      R = M - 1
    else:
      L = M + 1

  return False

binary_search(A, 8)
```

```python
# Condition based
B = [False, False, False, False, True]

def binary_search_condition(arr):
  N = len(arr)
  L = 0
  R = N - 1

  while L < R:
    M = (L + R) // 2

    if arr[M]:
      R = M
    else:
      L = M + 1

  return L

binary_search_condition(B)
```

```python
# Range-Based

def guess_number(target, L, R):
    while L <= R:
        M = (L + R) // 2
        if M == target:
            return M  # Found the number
        elif M < target:
            L = M + 1
        else:
            R = M - 1
    return -1  # Target not found


target = 2
result = guess_number(target, 1, 100)
print(f"The guessed number is: {result}")
```
