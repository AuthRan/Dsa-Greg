Like in this example whenever there is a collision by our hash function we have two ways to handle this collision -
1.Seperate Chaining : It simply means that each of the bucket in the hash table is going to be a linked list
2.Linear Probing : Abhi pdhaya noi gaya hai

But why do hashing even exist in first place, well the main point is to implement one of these things ie either a hash set(collection of unique items) or a hash map.
The most powerfull feature of hash is that it can look into the hash table ie search in a constant time ie O(1)

Well here is the tricky part that even though it is O(1) it is written as *O(1) so its amortized cost is constant because if u implemented hash table using Seperate chaining then even though to check the first node it is O(1) but if you have a chain of linked list at a particular hash table the time complexity will increase - Reason why choosing a good hash function is important.



Now talking hash maps it is basically all the functions of hash set plus you can store the data as in Key: value pairs. Keys must be hashable (we will look into that)
 
Now lets focus on the other technique to handle collisons that is linear probing: What we do here is that instead of creating chains of linked list at a fixed position in hash table we will assign the data to next empty spot after this spot but it does have problems when you delete bcz if you delete an element you might actually end up breaking linear probe structure. Ultimately this method is little weired and we do still get amortized cost of 1 in this case as well.


Now what is hashable?
Strings, Integer and Tuples (Sometimes are called frozen array)
What isn't?
Arrays/List, Dictionaries(even though they are hashmaps themselves)

Reason for these differences it becase all those who are hashable are immutable as mutable values can change that will cause inconsistency in hash table.

```python
# Hashsets

s = set()
print(s)
```

```python
# Add item into Set - O(1)
s.add(1)
s.add(2)
s.add(3)

print(s)
```

```python
# Lookup if item in set - O(1)
if 1 not in s:
  print(True)
```

```python
# Remove item from set - On average O(1)
s.remove(3)

print(s)
```

```python
# Set construction - O(S) - S is the length of the string
string = 'aaaaaaabbbbbbbbbbbccccccccceeeeeeeee'
sett = set(string)

sett
```

```python
# Loop over items in set - O(n) but very rarely used
for x in s:
  print(x)
```

```python
# Hashmaps - In python they are called dictionaries
# Here the values can be array, integer and string
d = {'greg': 1, 'steve': 2, 'rob': 3}
print(d)
```

```python
# Add key:val in dictionary: O(1)
d['arsh'] = 4

print(d)
```

```python
# Check for presence of key in dictionary: O(1)
if 'greg' in d:
  print(True)
```

```python
# Check the value corresponding to a key in the dictionary: O(1)
print(d['greg'])
# This will work only if greg is already preesnt in the dictionary otherwise it will throw an error.
```

```python
# Loop over the key:val pairs of the dictionary: O(n)
for key, val in d.items():
  print(f'key {key}: val {val}')
```

```python
# Now as hashmaps are so common in DSA so few clever ones exists like defaultdict
# Defaultdict

from collections import defaultdict

default = defaultdict(int)       
default2 = defaultdict(list)
# Unlike normal dict where u get an error if key doesn't exist it gives a default value back which in case of int is 0 or empty list in case of list
print(default[2])
default2
```

```python
# Counter
from collections import Counter

string = 'aaaaaaaaaaaabbbbbbbbbbbbbcccdddd'
counter = Counter(string)

print(counter)
# It is not recommended in interview to cheat it with rather best is go with either default dict or normal dictionary.
```
