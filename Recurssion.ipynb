Let's discuss a popular algorithm called recurssion. To illustrate this think of fibonacci numbers that has starting two values as 0 and 1 and later on all the values are sum of previous two values before it.


This tree is called recurssive tree. Now when a function is recurssive you will often times hear a term called "Call stack" stack because it uses stack data structure and call means function calls. So in recurssion you are calling the same function from inside the function this calling itself will generate a function call stack. Recursive functions requires a base case bcz other wise call stack will keep on growing.
 
The address that each call stack is storing is just the address it should returns the value if it has in future.

As you can probably guess the recurssive call here is f(n-2) + f(n-1) here the position of recursive call is important as the computer will first find the value of f(n-2) before he even thinks about f(n-1)


As you can see the space to calculate anything using recurssion would be the height of recursive tree. Also rem that the use of recurrsion is mostly seen in linked list as they are recurssive in nature (4->5->2->NULL) so their algorithms are also recurssive and trees are also recurssive as they have base case as leaf nodes that do not have any child and lastly graphs are also recussive in nature. In general you can solve many problems using these recurssion.

```python
# Fibonacci
# F(0) = 0
# F(1) = 1
# n > 1: F(n) = F(n-1) + F(n-2)

# Time: O(2^n), Space: O(n)
def F(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return F(n-1) + F(n-2)

F(12)
```

```python
# Linked Lists

class SinglyNode:

  def __init__(self, val, next=None):
    self.val = val
    self.next = next

  def __str__(self):
    return str(self.val)

Head = SinglyNode(1)
A = SinglyNode(3)
B = SinglyNode(4)
C = SinglyNode(7)

Head.next = A
A.next = B
B.next = C

print(Head)
```

```python
# Time: O(n), Space: O(n)
def reverse(node):
  if not node:
    return

  reverse(node.next)
  print(node)


reverse(Head)
```
